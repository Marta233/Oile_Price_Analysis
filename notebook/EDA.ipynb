### Boxplot
- The boxplot at the top shows the interquartile range (IQR) for dates in the dataset.

- The whiskers extend to the minimum and maximum dates, demonstrating that there are no significant outliers in the data distribution.
### Histogram

- There is no significant peak or drop in any specific period, which implies a balanced data collection over time.
### Mean and Median

## Data Visualization Insights
### 1. Boxplot
- The boxplot indicates the spread and skewness of the price data.
- The left side of the boxplot (25th percentile) is closer to the minimum, while the right side (75th percentile) extends further, suggesting a positive skew.
- There is a longer tail to the right, indicating the presence of higher price values as potential outliers or uncommon high values.
- The median (black line within the box) is visibly lower than the mean, reinforcing the positive skewness.
### 2. Histogram
- The histogram shows a large concentration of prices in the lower ranges, specifically between 10 and 30, with a sharp decrease afterward.
- There are smaller groups of values around higher price ranges (above 100), but these are less frequent.
- The distribution is right-skewed, meaning most prices are low, but there are some higher prices, which pull the mean (green dashed line) above the median (black line).
